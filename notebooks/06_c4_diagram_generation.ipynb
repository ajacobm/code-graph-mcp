## Section 1: Setup & Load Graph

```python
import os
import sys
import asyncio
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from typing import Dict, List, Set, Tuple
from collections import defaultdict, Counter

import httpx
import redis.asyncio as redis
from neo4j import GraphDatabase

plt.style.use('dark_background')
%matplotlib inline

print("âœ… Libraries loaded")
```

```python
# Initialize connections
import os
import socket

IN_CONTAINER = os.path.exists('/.dockerenv')
HOST_ALIAS = 'host.docker.internal' if IN_CONTAINER else 'localhost'

# Load environment variables with container-aware defaults
REDIS_URL = os.getenv('REDIS_URL', f'redis://{HOST_ALIAS}:6379')
MEMGRAPH_URL = os.getenv('MEMGRAPH_URL', f'bolt://{HOST_ALIAS}:7687')
BACKEND_API_URL = os.getenv('BACKEND_API_URL', f'http://{HOST_ALIAS}:10101')

class GraphConnections:
    def __init__(self, redis_url: str, memgraph_url: str, api_url: str):
        self.redis_url = redis_url
        self.memgraph_url = memgraph_url
        self.api_url = api_url
        self.redis_client = None
        self.memgraph_driver = None
        self.http_client = None
    
    async def connect(self):
        try:
            self.redis_client = await redis.from_url(self.redis_url, decode_responses=True)
            await self.redis_client.ping()
            print("âœ… Redis connected")
        except Exception:
            print("âš ï¸  Redis unavailable")
        
        try:
            self.memgraph_driver = GraphDatabase.driver(self.memgraph_url)
            with self.memgraph_driver.session() as session:
                session.run("RETURN 1").consume()
            print("âœ… Memgraph connected")
        except:
            print("âš ï¸  Memgraph unavailable")
            self.memgraph_driver = None
        
        self.http_client = httpx.AsyncClient(base_url=self.api_url)
        print("âœ… HTTP client ready")

connections = GraphConnections(REDIS_URL, MEMGRAPH_URL, BACKEND_API_URL)
loop = asyncio.new_event_loop()
asyncio.set_event_loop(loop)
await connections.connect())

print("\nâœ… Connected")
```

```python
# Load graph data
async def get_all_nodes():
    try:
        response = await connections.http_client.get('/api/graph/nodes/search?limit=1000')
        return pd.DataFrame(response.json().get('results', []))
    except Exception:
        return pd.DataFrame()

async def get_all_relationships():
    try:
        response = await connections.http_client.get('/api/graph/relationships?limit=5000')
        return response.json().get('results', [])
    except Exception:
        return []

nodes_df = await get_all_nodes()
relationships = await get_all_relationships()

# Build graph
G = nx.DiGraph()
if not nodes_df.empty:
    for idx, row in nodes_df.iterrows():
        G.add_node(row['name'], type=row.get('type'), file=row.get('file'), language=row.get('language'))

for rel in relationships:
    if rel.get('source_name') in G and rel.get('target_name') in G:
        G.add_edge(rel['source_name'], rel['target_name'])

print(f"âœ… Graph loaded: {G.number_of_nodes()} nodes")
```

## Section 2: Extract Modules (Level 2 Container)

Group functions into logical modules based on file organization.

```python
# Group nodes by file (module)
modules = defaultdict(list)
for idx, row in nodes_df.iterrows():
    file_path = row.get('file', 'unknown')
    # Extract top-level module (e.g., 'src/core/parser.py' -> 'core')
    if file_path and file_path != 'unknown':
        parts = file_path.split('/')
        if 'src' in parts:
            idx = parts.index('src')
            module = parts[idx+1] if idx+1 < len(parts) else 'unknown'
        else:
            module = parts[0] if parts else 'unknown'
    else:
        module = 'unknown'
    
    modules[module].append(row['name'])

# Show modules
module_sizes = {m: len(nodes) for m, nodes in modules.items()}
module_sizes = dict(sorted(module_sizes.items(), key=lambda x: x[1], reverse=True))

print("ðŸ“¦ Module/Container Analysis:\n")
for module, count in list(module_sizes.items())[:15]:
    print(f"  {module:30s}: {count:3d} entities")

# Visualize
fig, ax = plt.subplots(figsize=(12, 6))
modules_to_plot = dict(list(module_sizes.items())[:12])
ax.barh(list(modules_to_plot.keys()), list(modules_to_plot.values()), color='#4F46E5')
ax.set_xlabel('Entity Count', fontsize=12)
ax.set_ylabel('Module', fontsize=12)
ax.set_title('Top Modules/Containers', fontsize=14, fontweight='bold')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

## Section 3: Calculate Module Relationships

Determine dependencies between modules.

```python
# Map nodes to modules
node_to_module = {}
for module, nodes in modules.items():
    for node in nodes:
        node_to_module[node] = module

# Calculate inter-module dependencies
module_dependencies = defaultdict(set)
for rel in relationships:
    source = rel.get('source_name')
    target = rel.get('target_name')
    
    if source in node_to_module and target in node_to_module:
        source_module = node_to_module[source]
        target_module = node_to_module[target]
        
        if source_module != target_module:
            module_dependencies[source_module].add(target_module)

# Build module graph
G_modules = nx.DiGraph()
for module in modules.keys():
    G_modules.add_node(module, size=module_sizes.get(module, 1))

for source, targets in module_dependencies.items():
    for target in targets:
        G_modules.add_edge(source, target)

print(f"ðŸ”— Module Dependency Graph:")
print(f"   Modules: {G_modules.number_of_nodes()}")
print(f"   Dependencies: {G_modules.number_of_edges()}")

# Show dependencies
print(f"\nðŸ“Š Module Dependencies (top 10):\n")
out_degree = dict(G_modules.out_degree())
for module, count in sorted(out_degree.items(), key=lambda x: x[1], reverse=True)[:10]:
    targets = list(G_modules.successors(module))
    print(f"  {module} â†’ {', '.join(targets[:3])}{'...' if len(targets) > 3 else ''}")
```

## Section 4: Generate PlantUML C4 Diagram

Create PlantUML source code for C4 model visualization.

```python
# Generate PlantUML C4 diagram
def generate_c4_diagram(modules_dict: Dict, module_deps: Dict, project_name: str = "Code Graph") -> str:
    """
    Generate PlantUML C4 diagram (Level 2: Container)
    """
    lines = []
    lines.append("@startuml C4_System")
    lines.append("!include https://raw.githubusercontent.com/plantuml-stdlib/C4-PlantUML/master/C4_Container.puml")
    lines.append("")
    lines.append(f"title {project_name} - Container Diagram (C4 Level 2)")
    lines.append("")
    
    # System boundary
    lines.append(f"System_Boundary(sys, \"{project_name}\") {{")
    
    # Containers (modules)
    container_ids = {}
    sorted_modules = sorted(modules_dict.items(), key=lambda x: x[1], reverse=True)[:10]
    
    for i, (module, count) in enumerate(sorted_modules):
        container_id = f"c{i+1}"
        container_ids[module] = container_id
        tech = "Module"
        lines.append(f"    Container({container_id}, \"{module}\", \"{tech}\", \"{count} entities\")")
    
    lines.append("}")
    lines.append("")
    
    # Relationships
    rel_set = set()
    for source, targets in module_deps.items():
        if source in container_ids:
            for target in targets:
                if target in container_ids:
                    rel_key = tuple(sorted([source, target]))
                    if rel_key not in rel_set:
                        lines.append(f"Rel({container_ids[source]}, {container_ids[target]}, \"depends on\")")
                        rel_set.add(rel_key)
    
    lines.append("")
    lines.append("@enduml")
    
    return "\n".join(lines)

# Generate diagram
puml_c4 = generate_c4_diagram(module_sizes, module_dependencies, "code-graph-mcp")

print("âœ… PlantUML C4 Diagram Generated")
print("\n" + puml_c4)
```

## Section 5: Generate Mermaid Diagram

Create Mermaid syntax diagram for alternative rendering.

```python
# Generate Mermaid diagram
def generate_mermaid_diagram(modules_dict: Dict, module_deps: Dict) -> str:
    """
    Generate Mermaid graph diagram
    """
    lines = []
    lines.append("graph TD")
    lines.append("")
    
    # Add top modules
    sorted_modules = sorted(modules_dict.items(), key=lambda x: x[1], reverse=True)[:10]
    
    for module, count in sorted_modules:
        safe_name = module.replace('-', '_')
        lines.append(f"    {safe_name}[\"ðŸ“¦ {module}<br/>{count} entities\"]")
    
    lines.append("")
    
    # Add relationships
    for source, targets in module_deps.items():
        for target in targets:
            if source in module_sizes and target in module_sizes:
                safe_src = source.replace('-', '_')
                safe_tgt = target.replace('-', '_')
                lines.append(f"    {safe_src} -->|depends| {safe_tgt}")
    
    return "\n".join(lines)

# Generate diagram
mermaid_diagram = generate_mermaid_diagram(module_sizes, module_dependencies)

print("âœ… Mermaid Diagram Generated")
print("\n" + mermaid_diagram)
```

## Section 6: Visualize Module Architecture

Create networkx visualization of module dependencies.

```python
# Visualize module graph
plt.figure(figsize=(16, 12))

# Layout
pos = nx.spring_layout(G_modules, k=2, iterations=50, seed=42)

# Node sizes based on entity count
node_sizes = [module_sizes.get(node, 1) * 200 for node in G_modules.nodes()]

# Node colors based on out-degree (how many dependencies)
node_colors = [G_modules.out_degree(node) for node in G_modules.nodes()]

# Draw edges
nx.draw_networkx_edges(G_modules, pos, edge_color='gray', alpha=0.5, 
                        arrows=True, arrowsize=20, arrowstyle='->', width=2)

# Draw nodes
nodes = nx.draw_networkx_nodes(G_modules, pos, node_color=node_colors, 
                               node_size=node_sizes, cmap='YlOrRd', alpha=0.9)

# Draw labels
nx.draw_networkx_labels(G_modules, pos, font_size=10, font_weight='bold')

plt.colorbar(nodes, label='Number of Dependencies', shrink=0.8)
plt.title('Module Dependency Architecture (C4 Level 2)', fontsize=16, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()

print("âœ… Module architecture visualization complete")
```

## Section 7: Export Diagrams

Save generated diagrams to files.

```python
# Export diagrams
import os

output_dir = '/tmp/c4_diagrams'
os.makedirs(output_dir, exist_ok=True)

# Export PlantUML
with open(f'{output_dir}/c4_container_diagram.puml', 'w') as f:
    f.write(puml_c4)
print(f"âœ… PlantUML diagram exported: {output_dir}/c4_container_diagram.puml")

# Export Mermaid
with open(f'{output_dir}/module_dependencies.mmd', 'w') as f:
    f.write(mermaid_diagram)
print(f"âœ… Mermaid diagram exported: {output_dir}/module_dependencies.mmd")

# Export module data as JSON
import json
module_data = {
    "modules": module_sizes,
    "dependencies": {k: list(v) for k, v in module_dependencies.items()},
    "statistics": {
        "total_modules": len(modules),
        "total_dependencies": G_modules.number_of_edges(),
        "avg_module_size": np.mean(list(module_sizes.values()))
    }
}

with open(f'{output_dir}/module_data.json', 'w') as f:
    json.dump(module_data, f, indent=2)
print(f"âœ… Module data exported: {output_dir}/module_data.json")

print(f"\nðŸ“‚ All diagrams available in: {output_dir}")
```
